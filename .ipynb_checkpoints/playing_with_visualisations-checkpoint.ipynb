# Playing with seaborn

## Setup

### Importing libraries

```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Loading data

```python
# A330_year-count.csv
a330in = os.path.join("output", "processed_data", "A330_year-count.csv")
a330 = pd.read_csv(a330in)
```

```python
a330 = a330.rename(columns = {"0":"a330_count"})
```

```python
# A330_year.csv
a330fullIn = os.path.join("output", "processed_data", "A330_year.csv")
a330full = pd.read_csv(a330fullIn)
```

```python
# allAircraft_year-count.csv
allYearIn = os.path.join("output", "processed_data", "allAircraft_year-count.csv")
allYear = pd.read_csv(allYearIn)
```

```python
allYear.rename(columns = {"0":"count"}, inplace = True)
```

## Playing with plots

```python
sns.regplot(x="year", y="a330_count", data=a330)
```

```python
sns.lmplot(x="year", y="a330_count", data=a330)
```

```python
np.random.seed(1)
num_var = np.random.randn(1000)
num_var = pd.Series(num_var, name = "Numerical Variable")
```

```python
sns.histplot(data = num_var, kde = True)
```

## Testing with data

### Histogram with kde

```python
sns.histplot(data = a330full, x = "year", bins = 30, kde = True)
```

```python
a330full["year"].nunique()
```

## Trying to use allAircraft dataframe

```python
allIn = os.path.join("output", "processed_data", "allAircraft_datetime_year.csv")
All = pd.read_csv(allIn)
```

### Histogram with kde, hue=type

```python
sns.histplot(data = All, x = "year", kde = True, hue = "type")
```

### Kde-plot, hue=type

```python
sns.kdeplot(data = All, x = "year", hue = "type")
```

### Lineplot (needs year-count.csv files)

```python
sns.lineplot(data = a330, x = "year", y = "a330_count")
```

```python
sns.lineplot(data=allYear, x="year", y="count", hue="type", 
            palette="flare")
```

### Histogram, heatmap

```python
sns.histplot(All, x = "type", y = "year")
```

### Countplot (too year observations for this)

```python
sns.countplot(x = a330full["year"])
```

### Boxplot

```python
sns.catplot(data=All, x="type", y="year", kind="boxen")
```

### Violin plot

```python
sns.catplot(
    data=All, x="type", y="year",
    kind="violin", bw=.25
)
```

### lmplot, polynomial regression model

```python
sns.lmplot(x="year", y="a330_count", data=a330,
          order=2)
```

```python
sns.lmplot(x="year", y="count", data=allYear, 
           order=2, hue="type")
```

## Conclusion

- Use lineplot for combined table with the allAircraft dataframe.
- Use lineplot for a simple representation of the smaller dataframes.
- Use histogram for the smaller dataframes.
- Violin-plot or boxen-plot could be used to show longevity of some types over others.
- lmplot could be used to show tendency.
